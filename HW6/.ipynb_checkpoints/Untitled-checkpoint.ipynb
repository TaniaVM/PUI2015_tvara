```python
#nan are float, then we have to change the data to float
mta1 = mta1.astype(float)
mta1.dtype
```

```python
#Create a function to replace a number with nan
def replaceNoData(scanBlock, NDV):
    for n1, sB1 in enumerate(scanBlock):
        for  n2, sB2 in enumerate(sB1):
            for n3, elem in enumerate(sB2):
                if elem == NDV:
                    scanBlock[n1][n2][n3] = np.nan
replaceNoData(mta1,-1)
#mta2 = where(mta1<0, 'NaN', mta1)
```

```python

#Grouping all the elements of the 600 swipes and the 23 types of cards
#Adding all of them in one sum per week to identify the big drop
events = np.nansum(mta1,axis=(0,1))
```

```python
#Identifying exactly  (>3-sigma) in all time series.
```

```python
#Create an upper and lower thresholds, mean + 3 sigma 
upperthresholds = events.mean() + 3*events.std()
lowerthresholds = events.mean() - 3*events.std()
#Now Calculate Period of Time where it occured
Lowweeks = np.where(events < lowerthresholds) 
print '%s is the mean of all time series' % (events.mean())
print '%s is the standard deviation of all time series' % (events.std())
print '%s is the date that is below 3 standard deviations '% weeks[Lowweeks[0][0]]
```

Task 3: Periodicity
Identify the 4 stations (indentify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks).

```python
#Using Fourier Transform to know how much each periodicity contribute to the time series
perStations = np.nansum(mta1, axis=1)
yfreq = np.abs(np.fft.rfft(perStations))
#weeks periodicity 
xfreq = np.fft.rfftfreq(194, d=1.0)
#find the column that is closer to frequency 1/52 (annual)
idxAnnual = np.where(min(abs(xfreq - 1.0/52)) == abs(xfreq - 1.0/52))[0][0]
print idxAnnual
```

```python
#Find the 4 with bigger values
select1 = np.argpartition(yfreq[:,idxAnnual], -4)[-4:]
print select1
fig, ax = plt.subplots(nrows=2, ncols=2)
plt.figure(figsize(15,10))
for n,i in enumerate(select1):
    ax = fig.add_subplot(2,2,n+1)
    f = np.abs(np.fft.rfft(perStations[i,:])).transpose()
    ax.set_title("Station No %s" %(i))
    ax.plot(np.fft.rfftfreq(194, 1.0)[1:], (f)[1:], 'm-') 
    ax.plot(np.fft.rfftfreq(194, 1.0)[1:], (f)[1:], 'o')
    ax.set_xscale('log')
    ax.axvline(1.0/52)
    ax.set_xlabel('Frecuency (cycles/week)')
plt.subplots_adjust(wspace=.5, hspace=.5, left=.125, right=.9, top=1, bottom=.1)
```
