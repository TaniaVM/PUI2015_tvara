```python
import numpy as np
import pylab as pl
%pylab inline
```

```python
md={}
mymean = 100
mysize=1999/(np.array(range(1,100)+[10]))
```

```python
md={'chisq':{},'normal':{},'exponential':{},'gamma':{},'binomial':{},'poisson':{}}

for n in mysize:
    
    md['chisq'][n] = np.random.chisquare(mymean, size=n)
    md['normal'][n] = np.random.normal(mymean,10,n)
    md['exponential'][n] = np.random.exponential(mymean,n)
    md['gamma'][n] = np.random.gamma(mymean,1,n)
    md['binomial'][n] = np.random.binomial(2*mymean,0.5,n)
    md['poisson'][n] = np.random.poisson(mymean,n)
```

```python
means = {}
means ={'chisq':{},'normal':{},'exponential':{},'gamma':{},'binomial':{},'poisson':{}}

fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(1,1,1)
axnormal_mu_n = fig_mu_N.add_subplot(1,1,1)

for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    #and plot it
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
```
